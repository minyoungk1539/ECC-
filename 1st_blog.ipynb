```python
def isEmpty():
    return front == rear

def isFull():
    return rear == MAX_QSIZE

def enqueue(item):
    global rear
    if isFull():
        print("queue is full")
        return
    queue[rear] = item
    rear += 1

def dequeue():
    global front
    if isEmpty():
        print("queue is empty")
        return None
    item = queue[front]
    front += 1
    return item

def getFront():
    if isEmpty():
        return None
    return queue[front]

def getRear():
    if isEmpty():
        return None
    return queue[rear - 1]
```
